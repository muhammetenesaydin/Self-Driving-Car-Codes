```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# Rastgele sayı üreteci için seed değerini ayarlayarak, aynı sonuçları
# elde etmeyi garanti altına alıyoruz.
np.random.seed(0)

# Oluşturulacak veri noktalarının sayısını belirliyoruz.
n_pts = 500

# İki daire şeklinde bir veri seti oluşturuyoruz. 
# `n_samples`: Veri noktalarının sayısı.
# `random_state`: Rastgeleliği kontrol etmek için seed değeri.
# `noise`: Veri noktalarına eklenen gürültü miktarı.
# `factor`: İç ve dış dairelerin yarıçap oranı.
X, y = datasets.make_circles(n_samples=n_pts, random_state = 123, noise=0.1, factor=0.2)

# Oluşturulan veri noktalarını (X) ve etiketleri (y) yazdırıyoruz.
print(X) 
print(y)
 # Etiketleri 0 olan veri noktalarını mavi renkte çizdiriyoruz.
plt.scatter(X[y==0, 0], X[y==0, 1])
# Etiketleri 1 olan veri noktalarını turuncu renkte çizdiriyoruz.
plt.scatter(X[y==1, 0], X[y==1, 1])
```

```python
model = Sequential()
model.add(Dense(4, input_shape=(2,), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'],
    run_eagerly=False,
    steps_per_execution=1
)
h=model.fit(
    x=X,y=y,
    batch_size=20,
    epochs=100,
    verbose=1,
    callbacks=None,
    validation_split=0.0,
    validation_data=None,
    shuffle=True,
    class_weight=None,
    sample_weight=None,
    initial_epoch=0,
    steps_per_epoch=None,
    validation_steps=None,
    validation_batch_size=None,
    validation_freq=1
)



```

```python
plt.plot(h.history['accuracy'])
plt.legend(['accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
```

```python
plt.plot(h.history['loss'])
plt.legend(['loss'])
plt.title('loss')
plt.xlabel('epoch')
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:,0]) - 0.25, max(X[:,0]) + 0.25)
    y_span = np.linspace(min(X[:,1]) - 0.25, max(X[:,1]) + 0.25)
    xx, yy = np.meshgrid(x_span, y_span)
    grid = np.c_[xx.ravel(), yy.ravel()]
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
plot_decision_boundary(X, y, model)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plot_decision_boundary(X, y, model)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
 
x = 0
y = 0.75
 
point = np.array([[x, y]])
predict = model.predict(point)
plt.plot([x], [y], marker='o', markersize=10, color="red")
print("Prediction is: ", predict)
```
